```python
name = [i.get_full_name() for i in users1]
gender =  [i.get_gender() for i in users1]
location  = [i.get_city() for i in users1]
email = [i.get_email() for i in users1]
login =  [i.get_login_md5() for i in users1]
dob  = [i.get_dob() for i in users1]
registered = [i.get_registered() for i in users1]
phone = [i.get_phone() for i in users1]
picture = [i.get_picture() for i in users1]  # using for loop go over the data and collect everything we needed
```
